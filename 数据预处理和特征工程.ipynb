```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler, MaxAbsScaler, RobustScaler, Normalizer
from sklearn.impute import SimpleImputer
```

##### 数据无量纲化：将不同量纲（单位）的数据转换到相同量纲的过程，以便消除特征之间由于度量尺度不同带来的影响。无量纲化通常用于在机器学习模型中预处理数据，特别是在涉及距离计算或梯度下降的算法中，如支持向量机（SVM）、K-means聚类、线性回归和神经网络等。






```python
data = [[-1, -2], [-0.5, 6], [0, 10], [1, 18]]
```

```python
# 实现归一化
minmaxscaler = MinMaxScaler(
    feature_range = [5, 10] # 设置归一化区间
) # 实例化
minmaxscaler = minmaxscaler.fit(data) # fit
result = minmaxscaler.transform(data) # 导出结果

result= minmaxscaler.fit_transform(data) # inverse_transform
result
```
