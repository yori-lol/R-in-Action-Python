```python
# 全子集回归：所有可能的模型都会被检验，计算成本很高，特别是特征向量数量特别多的时候
from itertools import combinations
from sklearn.metrics import mean_squared_error
import statsmodels as 

X = states_[['Population', 'Income', 'Illiteracy', 'Life Exp', 'HS Grad', 'Frost', 'Area']]
Y = states_[['Murder']]
X, Y
```

```python
def all_subset_regression(X, Y):
    # 特征数量
    n_features = X.shape[1]

    best_score = float('inf')
    best_subset = None
    results = []

    for k in range(1, n_features + 1):
        for subset in combinations(X.columns, k): # 输出特征的排列组合
            model = sm.OLS(Y, sm.add_constant(X[list(subset)])).fit() # 理解sm.add_contant(X[list(subset)])
            score = mean_squared_error(Y, model.predict(sm.add_constant(X[list(subset)]))) # 如果判断条件调整为R2
            results.append((subset, score))
            if score < best_score:
                best_score = score
                best_subset = subset
    return best_score, best_subset, results
```

```python
for subset in combinations(X.columns, 2):
    print(subset)
```

```python
best_score, best_subset, results = all_subset_regression(X, Y)
best_score, best_subset
```

```python
# 几个指标
# MSE（均方误差）：残差的平方和 / 样本量大小，理论上，MSE越接近0表示模型的预测值与真实值越接近，但是存在过拟合问题
# RMSE（均方根误差）：MSE的开方
# MAE（平均绝对误差）：残差的绝对值之和 / 样本量大小
# R2 （确定系数）：1 - 残差的平方和 / 总平方和
```
